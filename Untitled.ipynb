```python
import pandas as pd
```

```python
df = pd.read_csv("./data/screenshots_infos.csv")
```

```python
df.sample(2)
```

```python
def get_valuemap_fname(s: pd.Series) -> str:
    return s.final_fname.replace("screen", "values").replace("png", "npy")

df["valuemap_fname"] = df.apply(get_valuemap_fname, axis=1)
```
